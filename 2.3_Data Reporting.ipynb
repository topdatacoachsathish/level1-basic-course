# Data Reporting From CSV


## Tabular Report
This code reads the CSV data and displays it as a tabular report using to_string(). You can customize the output further using pandas formatting 

```python
import pandas as pd

data = pd.read_csv("Basic_data.csv")

# Display the DataFrame (tabular report)
print(data.to_string())
```

## Chart Report
This code creates a bar chart showing the number of accounts for each gender using value_counts() and plot(). You can customize the chart title, labels, and colors. 

```python
# Count accounts by gender and plot a bar chart
gender_counts = data["Gender"].value_counts()
gender_counts.plot(kind="bar", color=["skyblue", "lightcoral"])

# Customize and display the chart (using matplotlib.pyplot)
import matplotlib.pyplot as plt
plt.title("Account Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

## Timeline Report (Line Chart by Account Opening Date)
This code converts the "AccOpen" column to datetime format, calculates the year, groups data by year, and calculates the average balance for each year. Finally, it creates a line chart showing the average balance trend over the years.

```python
# Convert 'AccOpen' to datetime format
data["AccOpen"] = pd.to_datetime(data["AccOpen"])

# Calculate the year for each account opening
data["Year"] = data["AccOpen"].dt.year

# Group data by year and calculate average balance
avg_balance_by_year = data.groupby("Year")["Balance"].mean()

# Plot line chart to show average balance over years
avg_balance_by_year.plot(kind="line", marker="o", color="green")

# Customize and display the chart
plt.title("Average Balance by Account Opening Year")
plt.xlabel("Year")
plt.ylabel("Average Balance")
plt.grid(True)
plt.show()
```

## Scatter Plot Report (Balance vs. Age)
This code creates a scatter plot to visualize the relationship between account holder's age and their account balance.

```python
# Scatter plot of balance vs. age
plt.figure(figsize=(8, 6))
plt.scatter(data["Age"], data["Balance"], alpha=0.7, color="purple")

# Customize and display the chart
plt.title("Balance vs. Age")
plt.xlabel("Age")
plt.ylabel("Balance")
plt.grid(True)
plt.show()
```
